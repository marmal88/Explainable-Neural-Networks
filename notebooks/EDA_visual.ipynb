Pneumonia is an inflammatory condition of the lung affecting primarily the small air sacs known as alveoli. The black patches within the ribcage outlines the shape of the lungs. Pneumonia can be spotted through X-Ray images by the white spots in the lungs. 

There are two types of pneumonia: Bacteria and virus infections. 

Bacteria tend to aggressively attack one lobe or section of the lungs causing a specific area of inflammation to take over the cells that were filled with air. When a provider hears lung sounds that seem normal on one side but absent on the other, bacterial pneumonia is more likely. An X-ray will show one white condensed area or opacity with the other areas of the lung visualized as having normal air exchange.

Viruses affect both sides of the lungs by producing a more homogeneous inflammatory reaction that causes increased cellular debris and mucus where previously open lung pockets were present.

## What can be expect to detect:

Normal: 

    - Clear Visual of the lungs with the bronchioles (branch)

Bacteria: 

    - Clear visual of one side of the lungs
    - the other will have a certain section of the lung that has condensed white area (due to inflammation)

Virus: 

    - Lungs with no condensed white area
    - Homogenous shade of white across the whole lungs
    - Blurred visuals of the bronchioles.
